```python
import requests
import time
import pandas as pd
from pandas import DataFrame

key = "319afee5c0380f4a1aed2e3df48a799f"
```

```python
url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json"
dfall = DataFrame()
def scrap(start, end):
    global dfall
    for i in range(start, end):
        params = {
            "key": key,
            "targetDt" : i
        }
    
        resp = requests.get(url, params=params)
        data = resp.json()
    
        df = DataFrame(data['boxOfficeResult']['dailyBoxOfficeList'])
        dfall = dfall.append(df)
    
    dfall['salesAmt'] = pd.to_numeric(dfall['salesAmt'])
    dfall['audiCnt'] = pd.to_numeric(dfall['audiCnt'])
```

```python
def scrap_year(year):
    global dfall
    year = str(year)
    scrap(int(year+"0101"), int(year+"0132"))
    time.sleep(2)
    scrap(int(year+"0201"), int(year+"0229"))
    time.sleep(2)
    scrap(int(year+"0301"), int(year+"0332"))
    time.sleep(2)
    scrap(int(year+"0401"), int(year+"0431"))
    time.sleep(2)
    scrap(int(year+"0501"), int(year+"0532"))
    time.sleep(2)
    scrap(int(year+"0601"), int(year+"0631"))
    time.sleep(2)
    scrap(int(year+"0701"), int(year+"0732"))
    time.sleep(2)
    scrap(int(year+"0801"), int(year+"0832"))
    time.sleep(2)
    scrap(int(year+"0901"), int(year+"0931"))
    time.sleep(2)
    scrap(int(year+"1001"), int(year+"1032"))
    time.sleep(2)
    scrap(int(year+"1101"), int(year+"1131"))
    time.sleep(2)
    scrap(int(year+"1201"), int(year+"1232"))
    
    dfall = dfall.drop(['rnum', 'rank', 'rankInten', 'rankOldAndNew', 'movieCd', 'openDt', 'salesShare', 'salesInten', 'salesChange', 'salesAcc', 'audiInten', 'audiChange', 'audiAcc', 'scrnCnt','showCnt'], axis=1)
    dfall.columns = ['제목', '매출액', '관객수']
    dfall = dfall.pivot_table(dfall, index = ['제목'], aggfunc = 'sum')
    
    dfall = dfall.sort_values(by=['관객수'], axis=0, ascending=False)
    dfall = dfall.reset_index()
```

```python
for i in range(2020, 2009, -1):
    scrap_year(i)
    dfall.to_csv(['trend_{}_2.csv'.format(i)], encoding="cp949")
    dfall[0:10].to_csv(['trend_{}.csv'.format(i)], encoding="cp949")
```
